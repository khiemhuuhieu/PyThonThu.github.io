```python
import pandas as pd
df = pd.read_csv('my_sales_data.csv', encoding='unicode_escape', header=0, delimiter=',')
df.head()
```

```python
dfEDWOrdered = df.groupby(['ORDERNUMBER']).size().reset_index(name='TotalOrderLineOrdered')
print(dfEDWOrdered)
```

```python
#(TotalProductOrdered) Số lượng hàng trên hóa đơn (tổng số lượng sản phẩm trên hóa đơn)
dfTemp = df.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].sum().reset_index(name='TotalProductOrdered')
dfTemp
```

```python
# merge
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
#(TotalSaleOrdered) Tổng giá trị hóa đơn đã bán
dfTemp = df.groupby(['ORDERNUMBER'])['SALES'].sum().reset_index(name='TotalSaleOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
# (AvgSaleProductTypeOrdered) Trung bình giá trị trên mỗi loại sản phẩm
dfTemp = df.groupby(['ORDERNUMBER'])['SALES'].mean().reset_index(name='AvgSaleProductTypeOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
# (AvgSaleEachProductOrdered) Trung bình giá trị trên mỗi sản phẩm
dfEDWOrdered['AvgSaleEachProductOrdered'] = dfEDWOrdered['TotalSaleOrdered']/dfEDWOrdered['TotalProductOrdered']
dfEDWOrdered
```

```python
# (AvgSaleEachProductOrdered) Trung bình giá trị trên mỗi sản phẩm
dfEDWOrdered['AvgSaleEachProductOrdered'] = dfEDWOrdered['TotalSaleOrdered']/dfEDWOrdered['TotalProductOrdered']
dfEDWOrdered
```

```python
#(AvgPriceEachProductOrdered) Trung bình đơn giá trên mỗi sản phẩm
dfEDWOrdered['AvgPriceEachProductOrdered'] = dfEDWOrdered['TotalSaleOrdered']/dfEDWOrdered['TotalProductOrdered']
dfEDWOrdered
#dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
#dfEDWOrdered
```

```python
#(AvgPriceProductTypeOrdered) Trung bình đơn giá trên mỗi loại sản phẩm
dfEDWOrdered['AvgPriceProductTypeOrdered'] = dfEDWOrdered['AvgPriceEachProductOrdered']*dfEDWOrdered['TotalOrderLineOrdered']
dfEDWOrdered
```

```python
# Chỉ lấy những dòng không trùng theo ['ORDERNUMBER','MONTH_ID','QTR_ID','YEAR_ID','CUSTOMERNAME','CITY','STATE','COUNTRY','TERRITORY']
dfTemp = df[['ORDERNUMBER','MONTH_ID','QTR_ID','YEAR_ID','CUSTOMERNAME','CITY','STATE','COUNTRY','TERRITORY']].drop_duplicates(subset=['ORDERNUMBER','MONTH_ID','QTR_ID','YEAR_ID','CUSTOMERNAME','CITY','STATE','COUNTRY','TERRITORY'])
#dfTemp[['ORDERNUMBER','MONTH_ID','QTR_ID','YEAR_ID','CUSTOMERNAME','CITY','STATE','COUNTRY','TERRITORY']].reset_index()
#dfTemp

dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
#(AvgProductTypeOrdered) Trung bình số lượng trên mỗi loại sản phẩm
dfEDWOrdered['AvgProductTypeOrdered'] = dfEDWOrdered['TotalProductOrdered']/dfEDWOrdered['TotalOrderLineOrdered']
dfEDWOrdered
```

```python
#(MaxProductTypeOrdered) Số lượng mặt hàng mua nhiều nhất trên mỗi hóa đơn
dfTemp = df.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].max().reset_index(name='MaxProductTypeOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
#(MinProductTypeOrdered) Số lượng mặt hàng mua thấp nhất trên mỗi hóa đơn
dfTemp = df.groupby(['ORDERNUMBER'])['QUANTITYORDERED'].min().reset_index(name='MinProductTypeOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
#(MaxPriceOrdered)Đơn giá cao nhất trong hóa đơn
dfTemp = df.groupby(['ORDERNUMBER'])['PRICEEACH'].max().reset_index(name='MaxPriceOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```

```python
#(MinPriceOrdered) Đơn giá thấp nhất trong hóa đơn
dfTemp = df.groupby(['ORDERNUMBER'])['PRICEEACH'].min().reset_index(name='MinPriceOrdered')
dfEDWOrdered = dfEDWOrdered.merge(dfTemp)
dfEDWOrdered
```
